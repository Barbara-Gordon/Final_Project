```python
test = debateCOL.find()

for r in range(0,41):
    id = test[r]['_id']
    
    text = test[r]['text']
    
    
    #basic cleanup
    text = text.replace("\n"," ")
    text = text.replace("\s+"," ")
    
    #pulling measurements/features from indicoio
    political_IND = indicoio.political(text)
    sentiment_IND = indicoio.sentiment_hq(text)
    personality_IND = indicoio.personality(text)
    emotion_IND = indicoio.emotion(text)
    personas_IND = indicoio.personas(text)
    people_IND = []
    places_IND = []
    organizations_IND = []
    
    people = indicoio.people(text)
    places = indicoio.places(text)
    things = indicoio.organizations(text)
    
    for p in range(len(people)):
        people_IND.append(people[p]['text'])
    
    for p in range(len(places)):
        places_IND.append(places[p]['text'])
   
    for t in range(len(things)):
        organizations_IND.append(things[t]['text'])
    
    #update the db
    debateCOL.update_one({'_id': id}, {'$set': {'text': text} } )
    debateCOL.update_one({'_id': id}, {'$set': {'political_IND': political_IND}})
    debateCOL.update_one({'_id': id}, {'$set': {'sentiment_IND': float(sentiment_IND)}})
    debateCOL.update_one({'_id': id}, {'$set': {'personality_IND': personality_IND}})
    debateCOL.update_one({'_id': id}, {'$set': {'emotion_IND': emotion_IND}})
    debateCOL.update_one({'_id': id}, {'$set': {'personas_IND': personas_IND}})
    debateCOL.update_one({'_id': id}, {'$set': {'people_IND': people_IND}})
    debateCOL.update_one({'_id': id}, {'$set': {'places_IND': places_IND}})
    debateCOL.update_one({'_id': id}, {'$set': {'organizations_IND': organizations_IND}})   
    
```
